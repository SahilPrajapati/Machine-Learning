```python

import pandas as pd

df = pd.read_csv(r'E:\Machine Learning\Exercises\machine-learning-ex1\ex1\ex1data2.txt',names = ['House Size(Square feet)',"#Bedrooms",'Price'])
df
```

```python
import numpy as np

x1 = df['House Size(Square feet)'].tolist()
x2 = df['#Bedrooms'].tolist()
y = df['Price'].tolist()

x0 = [1  for i in range(len(x1))]

XT = np.array([x0,x1,x2])

print(XT)

X = np.transpose(XT)

print('\n',X)

Y = np.transpose(np.array([y]))
print('\n',Y)
```

```python
theta = ((np.linalg.inv(XT.dot(X))).dot(XT)).dot(Y)

print(theta)
```

```python
print((np.array([1,1650,3])).dot(theta))
```
