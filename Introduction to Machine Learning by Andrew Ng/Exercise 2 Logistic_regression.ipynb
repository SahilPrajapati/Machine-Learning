```python
import pandas as pd
import numpy as np

#READING THE FILE
df = pd.read_csv(r'E:\Machine Learning\Exercises\machine-learning-ex2\ex2\ex2data1.txt', names = ['Exam1 Score','Exam2 Score','Admitted?'])

df
```

```python

#CREATING LISTS OF EACH COLUMN
x1 = df['Exam1 Score']
x2 = df['Exam2 Score']
y = df['Admitted?']

#Creating X matrix
x0 = [1 for i in range(len(x1))]
X = np.array([x0,x1,x2]).transpose()

print(X)


#creating Y matrix
Y = np.array([y]).transpose()
print(Y)

m = len(x1)
```

```python
#Creating theta

theta = np.array([[-24],[0.2],[0.2]])

print (theta)
```

```python
#PLOTTING...
import matplotlib.pyplot as plt

#Extracting the negative data i,e. where y = 0
neg = df[df['Admitted?']==0]

#Extracting the positive data i,e. where y = 1
pos = df[df['Admitted?']==1]


#Plotting the graph
plt.plot(neg['Exam1 Score'],neg['Exam2 Score'],'ko',label = 'Not admitted',color = 'magenta')
plt.plot(pos['Exam1 Score'],pos['Exam2 Score'],'k+',label = 'Admitted',color = 'black')
plt.xlabel('Exam1 Score')
plt.ylabel('Exam2 Score')
plt.legend()
```

```python
#SIGMOID FUNCTION...
import numpy as np

def sigmoid(z):
    
    #IF z IA AN ARRAY
    if isinstance(z,np.ndarray):
        temp = np.zeros(z.shape)
       
        temp = 1/(1 + np.exp(-z))
        
        return temp
   
    #IF z IS A LIST
    elif isinstance(z,list):
        for i in range(len(z)):
            z[i] = 1/(1 + np.exp(-z[i]))
        return z
    
    #IF z IS A NO.   
    else:
        return 1/(1 + np.exp(-z))
                
            
    
a = np.array([[1,2],[0,0],[4,0]])
#print(a)
print(sigmoid(a))
#print(sigmoid(78.02469282 ))
#print(X)
#print(sigmoid(X))
```

```python
#COST FUNCTION AND GRADIENT DESCENT FOR lOGISTIC REGRESSION...

#calculating the cost function
def costFunction(theta,X,Y):
    #first term of a cost function ylog(h(x))
    firstTerm = ((Y.transpose()).dot(np.log(sigmoid(X.dot(theta)))))
    
    #second term of a cost function (1-y)log(1-h(x))
    secondTerm = ((1-Y).transpose()).dot(np.log(1-sigmoid((X.dot(theta)))))
    
    # returning the total cost
    return -(((firstTerm+secondTerm))/m)


print('For theta : ',theta,'\n Cost Function is : ',costFunction(theta,X,Y))

    
```

```python
#Calculating only the gradient of the cost
def gradient(theta,X,Y):
    theta = (((X.transpose()).dot(sigmoid(X.dot(theta))-Y)/m))
    
    return theta


print ('For theta : ',theta,'\n\n Gradient Cost is : \n',gradient(theta,X,Y))
```

```python
#Using advance optimize algorithms...
import scipy.optimize as op
t = op.fmin(costFunction,x0=theta, args=(X, Y), maxiter=100,full_output = True)

print(t[0])
print(t[0][0])
print(t[1])
```

```python
#BFGS advanced algorithms
from scipy.optimize import minimize
res = minimize(costFunction, theta,args = (X,Y), method='BFGS',options = {'maxiter':500,'disp':True })
print(X.shape)
print(Y.shape)
print(theta.shape)
```

```python
#Now calculating for updated value of theta for exam 1 score = 45 & exam 2 score = 85
print(sigmoid(t[0][0] + t[0][1]*45 + t[0][2]*85))
```

```python
#Now Plotting the two features and the decision boundary

#for plotting the two reatures
plt.plot(neg['Exam1 Score'],neg['Exam2 Score'],'ko',label = 'Not admitted',color = 'magenta')
plt.plot(pos['Exam1 Score'],pos['Exam2 Score'],'k+',label = 'Admitted',color = 'black')

#for plotting the decision boundary i.e, theta[0] + theta[1]x1 + theta[2]x2 = 0
lst2 = [-((t[0][0]) + (t[0][1]*i))/t[0][2] for i in x1 ]
plt.plot(x1,lst2,label = 'decision boundary')


plt.xlabel('Exam1 Score')
plt.ylabel('Exam2 Score')
plt.legend()
```

```python
#PREDICTING THE ACCURACY

#Applying the hypothesis on the training data set for calculated parameters(theta)
Hx = sigmoid(X.dot(np.array([[t[0][0]], [t[0][1]],[t[0][2]]])))

count = 0

for i in range(m):
    if Hx[i] >= 0.5 and Y[i] == 1:
        count+=1
    elif Hx[i] < 0.5 and Y[i] == 0:
        count+=1
        
print('Total no. of data set : ',m)
print('Accurately predicted : ',count)
print('Percentage prediction : ',(count*100)/m,' %')
```
